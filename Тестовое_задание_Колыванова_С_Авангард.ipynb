### Какие подгруппы товаров наиболее часто покупают за все время продаж (минимум 4 группы)?

```python
df_1 = df['Sub-Category'].value_counts(dropna=False).reset_index()
df_1.columns = ['Sub-Category', 'count']
df_1['%'] = (df_1['count']/df_1['count'].sum() * 100).round(1)
df_1 
```

```python
plt.subplots(figsize=(15,8))
ax = sns.barplot(x='Sub-Category',
                 y='%',
                 data=df_1,
                 palette='pastel')

ax.set_title('Доля продаж по подгруппам за весь период', fontsize=20)
ax.set_xlabel('Подгруппа', fontsize=15)
ax.set_ylabel('Доля(%)', fontsize=15)

for bar in ax.patches:
    ax.annotate(format(bar.get_height(), '.1f'),
                (bar.get_x() + bar.get_width() / 2, bar.get_height()),
                ha='center', va='bottom',
                size=12, xytext=(0, 5),
                textcoords='offset points')

plt.xticks(rotation=45)
plt.show()
```

Для этого задания необходимо разбить все покупки на энное количество групп Sale_group (Допустим маленькие продажи, средние и высокие) на основе Sales Сгруппировать данные на основе региона и группы продаж (Region, Sale_group). Определить основные тенденции и паттерны. Выделить наиболее «прибыльную» группу.

Разобьем покупки на 3 группы

```python
# Определение интервалов и меток
bins = [0, 15000.0, 30000.0, float('inf')]  # Интервалы
labels = ['Низкие', 'Средние', 'Высокие']  # Метки

# Применение pd.cut() для создания новой колонки 'Sale_group'
df['Sale_group'] = pd.cut(df['Sales'], bins=bins, labels=labels, right=False)
```

```python
df.head()
```

### Распределение продаж по регионам и группам продаж

```python
df_g1 = df.groupby('Sale_group')['ID'].count().reset_index()
df_g1['%count'] = (df_g1['ID']/df_g1['ID'].sum() * 100).round(2)
df_g1
```

В процентном соотношении количество продаж в группах не отличается значительно.Однако чуть больше "Средних" продаж-34.52%."Высокие" составляют 33.58%,а "Низкие" -31.9% соответственно.

```python
df_g1 = df.groupby(['Region'], as_index=False).agg(
    Sales=('Sales', 'sum'),
    ID=('ID', 'count')
).sort_values(by='Sales', ascending=False).reset_index(drop=True)
df_g1['Sales'] = round(df_g1['Sales'])
df_g1['%sales'] = (df_g1['Sales']/df_g1['Sales'].sum() * 100).round(2)
df_g1['%count'] = (df_g1['ID']/df_g1['ID'].sum() * 100).round(2)
df_g1
```

В разрезе регионов Наибольшее число продаж и сумма приходится на Калининград.Далее идут Владивосток,Урал и Москва.

```python
df_g2 = df.groupby(['Sale_group'], as_index=False).agg(
    Sales=('Sales', 'sum'),
    ID=('ID', 'count')
).sort_values(by='Sales', ascending=False).reset_index(drop=True)
df_g2['Sales'] = round(df_g2['Sales'])
df_g2['%sales'] = (df_g2['Sales']/df_g2['Sales'].sum() * 100).round(2)
df_g2['%count'] = (df_g2['ID']/df_g2['ID'].sum() * 100).round(2)
df_g2
```

При этом наибольшая сумма продаж прикодится на "высокие"продажи- 54,79%(123075547.0 руб.)

Сгруппируем данные по регионам и группам продаж.

```python
df_c = df.groupby(['Region','Sale_group'], as_index=False).agg(
    Sales=('Sales', 'sum'),
    ID=('ID', 'count')
).sort_values(by='Sales', ascending=False).reset_index(drop=True)
df_c['Sales'] = round(df_c['Sales'])
df_c['%sales'] = (df_c['Sales']/df_c['Sales'].sum() * 100).round(2)
df_c['%count'] = (df_c['ID']/df_c['ID'].sum() * 100).round(2)
df_c
```

```python
# Получаем уникальные регионы
regions = df_c['Region'].unique()

# Настраиваем размер графиков
num_regions = len(regions)
num_cols = 2
num_rows =2
fig, axs = plt.subplots(nrows=num_rows, ncols=num_cols, figsize=(15, 6 * num_rows))
axs = axs.flatten()

# Создаем круговую диаграмму для каждого региона
for i, region in enumerate(regions):
    # Фильтруем данные для текущего региона
    data = df_c[df_c['Region'] == region]
    
    # Создаем круговую диаграмму
    axs[i].pie(data['Sales'], labels=data['Sale_group'], autopct='%1.1f%%', startangle=140)
    axs[i].set_title(f'Процентное соотношение продаж в регионе: {region}')
plt.show()
```
